```python
groups = [x[0] for x in df1.select("SC_CODE").distinct().collect()]
```

```python
stocks_dict = {"sc_name":None, "sc_code": None, "min_val": None, "min_at": None, "max_value": None, "max_at":None,  "current_value": None, "date": None }
```

```python
stock_list = []
df1.cache()
for group in groups[:100]:
    temp = stocks_dict.copy()
    temp_df = df1.filter(df1.SC_CODE == group)
    max_val = df1.filter(df1.SC_CODE == group).agg({'CLOSE':'max'}).collect()[0][0]
    min_val = df1.filter(df1.SC_CODE == group).agg({'CLOSE':'min'}).collect()[0][0]
    current_date = df1.filter(df1.SC_CODE == group).agg({'closing_date':'max'}).collect()[0][0]
    temp['sc_name'] = df1.filter(df1.SC_CODE == group)[['SC_NAME']].collect()[0][0]
    temp['sc_code'] = group
    temp['min_val'] = min_val
    temp['min_at'] = df1.filter(df1.SC_CODE == group).filter(temp_df.CLOSE == min_val).collect()[0]['closing_date']
    temp['max_value'] = max_val
    temp['max_at'] = df1.filter(df1.SC_CODE == group).filter(temp_df.CLOSE == max_val).collect()[0]['closing_date']
    temp['current_value'] = df1.filter(df1.SC_CODE == group).filter(temp_df.closing_date == current_date).collect()[0]['CLOSE']
    temp['date'] = current_date
    stock_list.append(temp)
df1.unpersist()
```

```python
new_df = spark.createDataFrame(data=stock_list)
new_df.printSchema()
new_df.show(truncate=False)
```

```python
from pyspark.sql.functions import lit
from pyspark.sql.types import FloatType

def cal_extra(pandas_df):
    pandas_df = pandas_df.assign(three_yr_avg_val=pandas_df.CLOSE.median())
    pandas_df = pandas_df.assign(three_yr_min_val=float(pandas_df.CLOSE.min()))
    pandas_df = pandas_df.assign(three_yr_max_val=float(pandas_df.CLOSE.max()))
    pandas_df = pandas_df.assign(current_val=float(pandas_df['CLOSE'].tail(1)))
    return pandas_df

df1 = df1.withColumn('three_yr_avg_val', lit(None).cast(FloatType()))
df1 = df1.withColumn('three_yr_min_val', lit(None).cast(FloatType()))
df1 = df1.withColumn('three_yr_max_val', lit(None).cast(FloatType()))
df1 = df1.withColumn('current_val', lit(None).cast(FloatType()))
df1.groupby('SC_CODE').applyInPandas(cal_extra, schema=df1.schema).display()
```
